```python
# Galime pakeisti reiksmes stulpelyje International plan(map funkcija)
```

```python
d = {"No" : False, "Yes" : True}
df["International plan"] = df["International plan"].map(d)
df.head()
```

```python
#Taip pat ir Voice mail plan stulpeli
d = {"No" : False, "Yes" : True}
df["Voice mail plan"] = df["Voice mail plan"].map(d)
df.head()
```

```python
# Pasidarau nauja stulpeli total charge
```

```python
df["Total charge"] = (
    df["Total day charge"]
    +df["Total eve charge"]
    +df["Total night charge"]
    +df["Total intl charge"]
)
df.head()
```

```python
# Stulpelis total calls
df["Total calls"] = (
    df["Total day calls"]
    +df["Total eve calls"]
    +df["Total night calls"]
    +df["Total intl calls"]
)
df.head()
```

```python
#Sugrupuoju pagal kelias stulpeliu reiksmes (pvz mazejimo)
```

```python
df.sort_values(by=["Churn", "Total day charge"], ascending=[True, False]).head()
```

```python
skambuciu_reiksmes = ["Total day minutes", "Total eve minutes", "Total night minutes"]
df.groupby(["Churn"])[skambuciu_reiksmes].describe(percentiles=[])
```
